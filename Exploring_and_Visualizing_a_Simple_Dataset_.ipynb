```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the iris dataset
iris = sns.load_dataset("iris")
```

```python
# Dataset shape
print("Shape:", iris.shape)

# Column names
print("Columns:", iris.columns.tolist())

# First 5 rows
print(iris.head())
```

```python
# Basic info
print(iris.info())

# Descriptive statistics
print(iris.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Width")
plt.show()
```

```python
iris.hist(figsize=(10, 8), bins=20)
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
for col in iris.columns[:-1]:
    sns.histplot(data=iris, x=col, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title("Box Plots of Iris Features")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Petal Length by Species")
plt.show()
```

```python
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_iris = iris.select_dtypes(include=np.number)

plt.figure(figsize=(8, 6))
# Calculate correlation on the numeric DataFrame
sns.heatmap(numeric_iris.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
```

```python
g = sns.FacetGrid(iris, col="species")
g.map(sns.histplot, "petal_length")
plt.show()
```

```python
from pandas.plotting import andrews_curves
andrews_curves(iris, "species")
plt.title("Andrews Curves")
plt.show()
```
